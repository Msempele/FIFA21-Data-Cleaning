### 11. Contract column

```python
fifa["Contract"].dtype
```

```python
fifa["Contract"].unique()
```

```python
for index, row in fifa.iterrows():
  if "On Loan" in row["Contract"] or "Free" in row["Contract"]:
    print (row["Contract"])
```

```python
def extract_contract_info(contract):
  if contract == "Free" or "On Loan" in contract:
    start_date = np.nan
    end_date = np.nan
    contract_length = 0
  else:
    start_date, end_date = contract.split(" ~ ")
    start_year = int(start_date[:4])
    end_year = int(end_date[:4])
    contract_length = end_year - start_year
  return start_date, end_date, contract_length

# Apply function to Contract column and create new columns
new_cols = ["Contract Start", "Contract End", "Contract Length(years)"]
new_data = fifa["Contract"].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
  fifa.insert(loc=fifa.columns.get_loc("Contract")+1+i, column=new_cols[i], value=new_data[i])
```

```python
fifa.sample(5)
```

```python
fifa[["Contract", 'Contract Start', 'Contract End', 'Contract Length(years)']].sample(5)
```

```python
# Contract Categories
def categorize_contract_status(contract):
  if contract == "Free":
    return "Free"
  elif "On Loan" in contract:
    return "On Loan"
  else:
    return "Contract"

fifa.insert(fifa.columns.get_loc("Contract Length(years)")+1, "Contract Status", fifa["Contract"].apply(categorize_contract_status))
fifa.sample(3)
```

### 13. Height

```python
fifa["Height"].dtype
```

```python
fifa["Height"].unique()
```

```python
def convert_height(height):
  if "cm" in height:
    return int(height.strip("cm"))
  else:
    feet, inches = height.split("'")
    total_inches = int(feet) * 12 + int(inches.strip('"'))
    return round(total_inches * 2.54)

# Apply function to the height column
fifa["Height"] = fifa["Height"].apply(convert_height)
fifa["Height"].unique()
```

```python
fifa = fifa.rename(columns={"Height": "Height(cm)"})
fifa.sample(3)
```

### 14. Weight

```python
fifa["Weight"].dtype
```

```python
fifa["Weight"].unique()
```

```python
def convert_weight(weight):
  if "kg" in weight:
    return int(weight.strip("kg"))
  else:
    pounds = int(weight.strip("lbs"))
    total_kg = round(pounds * 0.453592)
    return round(total_kg)

# Apply function to Weight column
fifa['Weight'] = fifa['Weight'].apply(convert_weight)
fifa['Weight'].unique()
```

```python
# Rename column name
fifa = fifa.rename(columns={"Weight": "Weight(kg)"})
fifa.sample(3)
```

### 20. Value

Estimated market value of the player.

```python
fifa["Value"].dtype
```

```python
fifa["Value"].unique()
```

```python
def clean_value(value):
  value = value.replace("€", "")
  if "M" in value:
    return int(float(value.replace("M","")) * 1000000)
  elif "K" in value:
    return int(float(value.replace("K","")) * 1000)
  else:
    return int(value)

# Apple the function to the Value column
fifa["Value"] = fifa["Value"].apply(clean_value)
fifa["Value"].unique()
```

```python
# Rename Value column
fifa = fifa.rename(columns={"Value": "Value(€)"})
fifa.sample(3)
```

### 77. Hits

Number of times the player has been searched for in the FIFA database.

```python
fifa["Hits"].dtype
```

```python
fifa["Hits"].unique()
```

```python
def remove_k(hits):
  if pd.isna(hits):
    return (hits)
  hits = str(hits)
  if "K" in hits:
    return int(float(hits.replace("K", "")) * 1000)
  else:
    return int(float(hits))


# Apply function to Hits column
fifa["Hits"] = fifa["Hits"].apply(remove_k)
fifa["Hits"].unique()
```

```python

fifa["Hits"].unique()
```

```python
def convert_hits_col(hits_col):
  if pd.isna(hits_col):
    return (hits_col)
  else:
    return int(hits_col)

# Apply function to Hits column
fifa["Hits"] = fifa["Hits"].apply(convert_hits_col)
fifa["Hits"].unique()

```
